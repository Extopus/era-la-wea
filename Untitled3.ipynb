```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#realizamos las importaciones necesarias
```

```python
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
    name=fn
df = pd.read_csv(name, sep=",", encoding='latin1')
df.head(10)
#cargamos la base de datos y hacemos un ciclo for para visualizar las 10 primeras filas de la base de datos para darnos una idea de esta misma
```

```python



categorical_vars = ['InvoiceNo', 'StockCode', 'Description', 'Country', 'CustomerID']



numerical_vars = ['Quantity', 'UnitPrice']



print("Variables categóricas:")
print(categorical_vars)

print("\nVariables numéricas:")
numerical_vars
```

podemos ver las variables categoricas que como dice el nombre son para categorizar datos en grupos como ['InvoiceNo', 'StockCode', 'Description', 'Country', 'CustomerID'] y las variables numericas como ['Quantity', 'UnitPrice'] que representan cantidades y/o mediciones

```python

mean_price = df['UnitPrice'].mean()
median_price = df['UnitPrice'].median()
mode_price = df['UnitPrice'].mode()[0]
print(f"Mean UnitPrice: {mean_price}")
print(f"Median UnitPrice: {median_price}")
print(f"Mode UnitPrice: {mode_price}")
```

podemos visualizar la media, mediana y moda de la columna UnitPrice dandonos a entender que la media de precio de los productos es de 4.6, la mediana o el dato a la mitad de la columna es de 2.08 y el precio mas comun es la tabla es de 1.25.

```python

import matplotlib.pyplot as plt
product_sales = df.groupby('Description')['Quantity'].sum()


top_10_products = product_sales.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_products.index, y=top_10_products.values)
plt.xticks(rotation=90)
plt.xlabel('Product')
plt.ylabel('Total Quantity Sold')
plt.title('Top 10 Best-Selling Products')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['Month'] = df['InvoiceDate'].dt.month

plt.figure(figsize=(12, 6))
sns.histplot(df['Month'], bins=12, kde=False)
plt.xticks(range(1, 13))
plt.xlabel('Month')
plt.ylabel('Number of Purchases')
plt.title('Monthly Distribution of Purchases')
plt.show()
```

En este histograma podemos visualizar el comportamiento temporal de las compras en los meses de la base de datos, podemos ver que en el mes que mas se compra es en el mes 11, osea noviembre con mas de 80 mil compras, en cambio podemos ver tambien que el mes en el que menos se vende es en el mes 2, osea febrero con menos de 30 mil ventas en el mes.
