```python

import numpy as np
# Medidas de tendencia central
def medidas_tendencia_central(df, columna):
  media = df[columna].mean()
  mediana = df[columna].median()
  moda = df[columna].mode()[0]  # Tomamos el primer valor de la moda si hay varios
  print(f"Medidas de tendencia central para la columna '{columna}':")
  print(f"- Media: {media}")
  print(f"- Mediana: {mediana}")
  print(f"- Moda: {moda}")


# Medidas de dispersión
def medidas_dispersion(df, columna):
  rango = df[columna].max() - df[columna].min()
  varianza = df[columna].var()
  desviacion_estandar = df[columna].std()
  print(f"\nMedidas de dispersión para la columna '{columna}':")
  print(f"- Rango: {rango}")
  print(f"- Varianza: {varianza}")
  print(f"- Desviación estándar: {desviacion_estandar}")


# Ejemplo de uso (reemplaza 'nombre_columna' con el nombre real de tu columna numérica)
for col in df.select_dtypes(include=np.number).columns:
  medidas_tendencia_central(df, col)
  medidas_dispersion(df, col)
```

```python
# prompt: necesito Identificar los valores atípicos.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# ... (previous code) ...

def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Ejemplo de uso para la columna 'UnitPrice'
outliers_unit_price = identify_outliers(df, 'UnitPrice')
print("\nValores atípicos en 'UnitPrice':")
print(outliers_unit_price)


# Puedes visualizar los outliers con un boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['UnitPrice'])
plt.title('Boxplot de UnitPrice')
plt.show()

#Para otras columnas numéricas:
for col in df.select_dtypes(include=np.number).columns:
    outliers = identify_outliers(df,col)
    print(f"\nValores atípicos en '{col}':")
    print(outliers)
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot de {col}')
    plt.show()
```

```python


import matplotlib.pyplot as plt
# Filter the DataFrame
filtered_df = df[(df['Quantity'] >= 0) & (df['Quantity'] <= 100) & (df['UnitPrice'] >= 0) & (df['UnitPrice'] <= 50)]

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['Quantity'], filtered_df['UnitPrice'])
plt.xlabel('Quantity')
plt.ylabel('UnitPrice')
plt.title('Scatter Plot of Quantity vs. UnitPrice (Filtered)')
plt.xlim(0, 100)  # Set x-axis limit
plt.ylim(0, 50)  # Set y-axis limit
plt.grid(True)
plt.show()
```

```python

import matplotlib.pyplot as plt
product_sales = df.groupby('Description')['Quantity'].sum()


top_10_products = product_sales.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_products.index, y=top_10_products.values)
plt.xticks(rotation=90)
plt.xlabel('Product')
plt.ylabel('Total Quantity Sold')
plt.title('Top 10 Best-Selling Products')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['Month'] = df['InvoiceDate'].dt.month

plt.figure(figsize=(12, 6))
sns.histplot(df['Month'], bins=12, kde=False)
plt.xticks(range(1, 13))
plt.xlabel('Month')
plt.ylabel('Number of Purchases')
plt.title('Monthly Distribution of Purchases')
plt.show()
```

En este histograma podemos visualizar el comportamiento temporal de las compras en los meses de la base de datos, podemos ver que en el mes que mas se compra es en el mes 11, osea noviembre con mas de 80 mil compras, en cambio podemos ver tambien que el mes en el que menos se vende es en el mes 2, osea febrero con menos de 30 mil ventas en el mes.
